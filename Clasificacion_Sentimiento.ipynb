"Transformers" es una biblioteca que permite el uso de pre-entrenados modelos de NLP desarrollados por Hugging Face.

```python
# Importando Pipeline de Transformers
from transformers import pipeline
```

"Pipeline" es una herramienta que permite realizar tareas de NLP de forma rápida y sencilla utilizando los modelos pre-entrenados disponibles en la librería.

```python
# Inicialización de Modelo de Clasificación de Sentimiento
modelo = 'pysentimiento/robertuito-sentiment-analysis'
pipeline_de_clasificador = pipeline("text-classification",model=modelo)
```

```python
# Data de prueba y cuadrito para que se vea presentable
import pandas as pd

def clasificador(text):
    """
    Función que clasifica un texto utilizando un pipeline de clasificación.

    Parámetros:
        text (str): El texto a clasificar.

    Devuelve:
        str: La clase asignada al texto.
    """
    return pipeline_de_clasificador(text)[0]['label']

data_de_prueba = {'Texto': ['Estoy feliz 🤗 de mostrarles un toolkit para Análisis de Sentimientos y otras tareas de SocialNLP',
                  'Espero que no lo odien.',
                  'Lo odiamos.']}

df = pd.DataFrame(data_de_prueba)
df['Sentimiento'] = df['Texto'].apply(clasificador)

def extender_resultado(Sentimiento):
    """
    Función que extiende una etiqueta de Sentimiento a su formato completo.

    Parámetros:
        Sentimiento (str): La etiqueta de Sentimiento en formato abreviado.

    Devuelve:
        str: La etiqueta de Sentimiento en formato completo.
    """
    if Sentimiento == 'POS':
        return 'Positivo'
    elif Sentimiento == 'NEU':
        return 'Neutro'
    elif Sentimiento == 'NEG':
        return 'Negativo'

df['Sentimiento'] = df['Sentimiento'].apply(extender_resultado)
df
```
